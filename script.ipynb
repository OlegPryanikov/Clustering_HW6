```python
data = {'Регион': [], 
         'Количество договоров страхования': [],
         'Страховые суммы по договорам страхования': [],
         'Количество договоров страхования': [], 
         'Количество урегулированных страховых случаев': [], 
         'Выплаты по договорам страхования': [], 
         'Страховые премии': [], 
        }

df1 = pd.read_excel('Сведения о количестве договоров страхования, заключенных в отчетном периоде.xlsx')
df2 = pd.read_excel('Сведения о страховых суммах по договорам страхования, заключенным в отчетном периоде.xlsx')
df3 = pd.read_excel('Сведения о количестве договоров страхования, заключенных в отчетном периоде.xlsx')
df4 = pd.read_excel('Сведения о выплатах по договорам страхования.xlsx')
df5 = pd.read_excel('Сведения о страховых премиях (взносах) по договорам страхования.xlsx')


for i in range(11,104):
    if i != 30 and i != 42 and i != 51 and i != 59 and i != 74 and i != 81 and i != 92:
        data['Регион'].append(df1.iat[i,0])

for i in range(11,104):
    if i != 30 and i != 42 and i != 51 and i != 59 and i != 74 and i != 81 and i != 92:
        data['Количество договоров страхования'].append(df1.iat[i,9])

for i in range(11,104):
    if i != 30 and i != 42 and i != 51 and i != 59 and i != 74 and i != 81 and i != 92:
        data['Страховые суммы по договорам страхования'].append(df2.iat[i,9])

for i in range(11,104):
    if i != 30 and i != 42 and i != 51 and i != 59 and i != 74 and i != 81 and i != 92:
        data['Количество урегулированных страховых случаев'].append(df3.iat[i,9])      

for i in range(11,104):
    if i != 30 and i != 42 and i != 51 and i != 59 and i != 74 and i != 81 and i != 92:
        data['Выплаты по договорам страхования'].append(df4.iat[i,9])      

for i in range(11,104):
    if i != 30 and i != 42 and i != 51 and i != 59 and i != 74 and i != 81 and i != 92:
        data['Страховые премии'].append(df5.iat[i,9])      


DF = pd.DataFrame(data)
DF  
```
